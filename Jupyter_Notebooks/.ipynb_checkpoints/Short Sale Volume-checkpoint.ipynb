```python
df = pd.read_csv(data_location)
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df['shortvolume'] = df['totalvolume'] - df['shortvolume']
df['short_ratio'] = 1 - df['short_ratio']

df
```

```python
# Function to grab short ratio data for a given ticker 
def short_ratio(ticker = 'GME'):
    df_ = df[df['ticker'] == ticker].copy()
    df_['date'] = pd.to_datetime(df_['date'])
    df_.set_index('date', inplace = True)
    return df_
    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Function 1: Check if a stock's short ratio ever exceeds SPY's std deviation thresholds with rolling window
def check_exceeds_threshold(stock_df, spy_df, std_values, window):
    spy_mean = spy_df['short_ratio'].rolling(window).mean()
    spy_std = spy_df['short_ratio'].rolling(window).std()
    thresholds = [spy_mean + x * spy_std for x in std_values]
    
    for threshold in thresholds:
        if (stock_df['short_ratio'].rolling(window).mean() > threshold).any():
            return True
    return False

# Adjusted Function 2: Calculate the SPY std deviation thresholds with rolling window
def calculate_spy_thresholds(spy_df, std_values, window):
    rolling_mean = spy_df['short_ratio'].rolling(window).mean()
    rolling_std = spy_df['short_ratio'].rolling(window).std()

    # Generate thresholds based on the rolling window
    thresholds = {f'{x} STD': rolling_mean + (x * rolling_std) for x in std_values}
    return thresholds

# Adjusted Function 3: Matplotlib Visualization with Highlighted Areas
def mpl_stock_vs_spy(res_aligned, res_spy_aligned, thresholds = [1, 1.5], window = 30):
    
    plt.figure(figsize=(20, 10))
    
    # Ticker
    ticker = res_aligned['ticker'][0]
    
    # Assuming 'res_aligned' is your DataFrame and you have already calculated the rolling mean
    rolling_mean = res_spy_aligned['short_ratio'].rolling(window).mean()
    rolling_std = res_spy_aligned['short_ratio'].rolling(window).std()
    
    # Calculate the 1.5 and 2 standard deviation values from the rolling mean
    one_and_half_std_above = rolling_mean + thresholds[0] * rolling_std
    two_std_above = rolling_mean + thresholds[1] * rolling_std

    # Plot the SPY lines
    plt.plot(res_spy_aligned.index, one_and_half_std_above, label='{} STD SPY'.format(str(thresholds[0])), color='orange', linestyle='--')
    plt.plot(res_spy_aligned.index, two_std_above, label='{} STD SPY'.format(str(thresholds[1])), color='red', linestyle='--')

    # Plot the GME line
    plt.plot(res_aligned.index, res_aligned['short_ratio'].rolling(window).mean(), label='{} 30-day Rolling Mean'.format(ticker), color='blue')

    # Full height shaded regions
    exceeds_mask = res_aligned['short_ratio'].rolling(window).mean() > one_and_half_std_above
    plt.fill_between(res_aligned.index, 0, 1, where=exceeds_mask, color='green', alpha=0.3, transform=plt.gca().get_xaxis_transform(), label='{} Exceeds 1.5 STD SPY'.format(ticker))

    # Optionally, add a fill between the 1.5 STD and 2 STD lines for SPY to indicate the 'danger zone'
    plt.fill_between(res_spy_aligned.index, one_and_half_std_above, two_std_above, color='yellow', alpha=0.1, label='SPY {} to {} STD Zone'.format(str(thresholds[0]), str(thresholds[1])))

    # Customizing the plot
    plt.title('{} vs SPY Short Ratio with Standard Deviation Bands'.format(ticker))
    plt.xlabel('Date')
    plt.ylabel('Short Ratio')
    plt.legend()
    plt.grid(True)
    plt.ylim(res_aligned['short_ratio'].min(), two_std_above.max())  # Set y-limits to data range
    plt.show()



# Adjusted Function 4: Plotly Visualization with Highlighted Areas and Tooltips
def plotly_stock_vs_spy(res_aligned, res_spy_aligned, thresholds = [1, 1.25], window = 30, width=1300, height=750):
    
    # Ticker
    ticker = res_aligned['ticker'][0]
    
    # Assuming 'res_aligned' is your DataFrame and you have already calculated the rolling mean
    rolling_mean = res_spy_aligned['short_ratio'].rolling(window).mean()
    rolling_std = res_spy_aligned['short_ratio'].rolling(window).std()
    
    # Calculate the 1.5 and 2 standard deviation values from the rolling mean
    one_and_half_std_above = rolling_mean + thresholds[0] * rolling_std
    two_std_above = rolling_mean + thresholds[1] * rolling_std
    exceeds_mask = res_aligned['short_ratio'].rolling(window).mean() > one_and_half_std_above
    highlight_dates = res_aligned.index[exceeds_mask]
    
    # Create a figure with secondary y-axis
    fig = make_subplots(specs=[[{"secondary_y": True}]])

    # Add SPY standard deviation lines
    fig.add_trace(
        go.Scatter(x=res_spy_aligned.index, y=one_and_half_std_above, name='{} STD SPY'.format(str(thresholds[0])), line=dict(color='orange', dash='dash')),
        secondary_y=False,
    )

    fig.add_trace(
        go.Scatter(x=res_spy_aligned.index, y=two_std_above, name='{} STD SPY'.format(str(thresholds[1])), line=dict(color='red', dash='dash')),
        secondary_y=False,
    )

    # Add GME rolling mean line
    fig.add_trace(
        go.Scatter(x=res_aligned.index, y=res_aligned['short_ratio'].rolling(window).mean(), name='{} 30-day Rolling Mean'.format(str(ticker)), line=dict(color='blue')),
        secondary_y=False,
    )

    # Add vertical lines for highlighted dates
    for date in highlight_dates:
        fig.add_trace(
            go.Scatter(
                x=[date, date], 
                y=[0, 1], 
                mode="lines",
                name=str(date.date()),  # Convert timestamp to date and set as name
                hoverinfo='name',  # Only show the name (i.e., the date) in the tooltip
                showlegend=False,  # Hide these traces from the legend
                line=dict(color='green', width=2),  # Apply line properties here
                opacity=0.5  # Apply opacity to the whole trace
            ),
            secondary_y=True,  # This ensures the line spans the full height of the plot
        )


    # Customizing the plot
    fig.update_layout(
        title='{} vs SPY Short Ratio with Standard Deviation Bands'.format(str(ticker)),
        xaxis_title='Date',
        yaxis_title='Short Ratio',
        legend_title='Legend',
        hovermode='closest',  # Show tooltip for the closest point
        height = height,
        width = width
    )


    # Set the range of y-axis to match your data
    fig.update_yaxes(range=[res_aligned['short_ratio'].min(), two_std_above.max()])

    # Show the figure
    fig.show()





```

```python
# Define the rolling window size
rolling_window = 30

# Check if GME exceeds SPY's std deviation thresholds with rolling window
exceeds = check_exceeds_threshold(res_aligned, res_spy_aligned, [1.5, 2], rolling_window)
print(f"Does GME exceed SPY's thresholds? {exceeds}")

# Plot GME vs SPY with matplotlib using rolling window
mpl_stock_vs_spy(res_aligned, res_spy_aligned, [1.5, 2], rolling_window)

# Plot GME vs SPY with plotly using rolling window
plotly_stock_vs_spy(res_aligned, res_spy_aligned, [1.5, 2], rolling_window)
```

```python
# Get ticker list based on which tickers have sufficient data 

tickers = df.ticker.value_counts().where(lambda x: x > 1300).dropna().index.tolist()
```

```python
from tqdm import tqdm

# Initialize the dictionaries
std_1_dict = {}
std_1_5_dict = {}
std_2_dict = {}

# Define the function to check if a ticker exceeds the SPY thresholds and get dates
def get_dates_exceeding_spy_thresholds(ticker_df, spy_df, std_values, window):
    results = {}
    for std_value in std_values:
        # Calculate SPY thresholds
        spy_threshold = calculate_spy_thresholds(spy_df, [std_value], window)[f'{std_value} STD']
        # Check dates when ticker exceeds the threshold
        exceeds_dates = ticker_df['short_ratio'].rolling(window).mean() > spy_threshold
        dates = ticker_df.index[exceeds_dates].tolist()
        if dates:
            results[std_value] = dates
    return results

# Loop through each ticker and populate the dictionaries
spy_df = short_ratio('SPY')
for ticker in tqdm(tickers[:100]):
    # Get short sale data
    ticker_df = short_ratio(ticker)
    
    # Align dataframes
    ticker_df, spy_df = ticker_df.align(spy_df, axis=0)
    
    exceeding_dates = get_dates_exceeding_spy_thresholds(ticker_df, spy_df, [1, 1.5, 2], window=30)
    if 1 in exceeding_dates:
        std_1_dict[ticker] = exceeding_dates[1]
    if 1.5 in exceeding_dates:
        std_1_5_dict[ticker] = exceeding_dates[1.5]
    if 2 in exceeding_dates:
        std_2_dict[ticker] = exceeding_dates[2]

# At the end, std_1_dict, std_1_5_dict, and std_2_dict will contain the tickers and dates as required
```

```python
tic = 'AOR'
rolling_window = 30

# Get short sale data
ticker_df = short_ratio(tic)

# Align dataframes
ticker_df, spy_df = ticker_df.align(spy_df, axis=0)   

# Check if ticker exceeds SPY's std deviation thresholds with rolling window
exceeds = check_exceeds_threshold(ticker_df, spy_df, [1.5, 2], rolling_window)
print(f"Does ticker's exceed SPY's thresholds? {exceeds}")

# Plot GME vs SPY with matplotlib using rolling window
mpl_stock_vs_spy(ticker_df, spy_df, [1.5, 2], rolling_window)

# Plot GME vs SPY with plotly using rolling window
plotly_stock_vs_spy(ticker_df, spy_df, [1.5, 2], rolling_window)
```
